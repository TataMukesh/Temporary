```python
board = [['X','O','X'],['X','O','O'],['O','X','O']]
```

```python
#Printing the board

for row in board:        #Sub lists will be assigned into variable row
    for column in row:   #Contents of each sub list will be assigned into variable column
        print(column,end = ' ')   #Print each character and specify end = " "
    print()   #This print() is used to go to new line after printing one row
```

```python
#T count number of cells filled with either X or O

def count(board):
    entries = 0    #entires is a variable to keep track of count
    for i in range(3):     # iterate through rows
        for j in range(3):   # iterate through columns
            if board[i][j] == 'X' or board[i][j] == 'O':   #If cell is filled with either X or O
                entries += 1             #Increase the variable entries by 1
    return entries
```

```python
"""A game can be won row wise or column wise or diagonal wise. So we need to check threee cases here"""


def check_winner(board):
    #Checking if any player has won row wise
    
    for row in board:    #Iterate for every sublist
        if len(set(row)) == 1:    #Since length of set returns only number of unique elements. 
            if row[0] == 'O':
                return "Player 0 wins"
            elif row[0] == 'X':
                return "Player 1 wins"
            else:
                continue
                
                
    #Checking if any player has won column wise
    #Here we use a trick to check if player has won column wise
    #First we transpose the board and then check row wise.
    #Checking column wise and checking row wise after transposing the board are one and the same. . Think this once
    
    board_transpose = [['','',''],['','',''],['','','']]  #Create a new matrix to store the transpose
    # iterate through rows
    for i in range(len(board)):
   # iterate through columns
       for j in range(len(board[0])):
           board_transpose[j][i] = board[i][j]
    
    #after transposing the board check row wise
    for row in board_transpose:
        if len(set(row)) == 1:
            if row[0] == 'O':
                return "Player 0 wins"
            elif row[0] == 'X':
                return "Player 1 wins"
            else:
                continue
                
    #Checking if any player has won diagonal wise           
    
    diagonal1 = set([board[0][0],board[1][1],board[2][2]])   #Left Diagonal
    diagonal2 = set([board[0][2],board[1][1],board[2][0]])   #Right Diagonal
    if len(diagonal1) == 1 or len(diagonal2) == 1 and board[1][1] != '':
        if board[1][1] == 'O':
            return "Player 0 wins"
        else:
            return "Player 1 wins"
        
    #If all the entries are filled with either X or O and there is no winner case
        
    if count(board) == 9:
        return "No one has won the game....It's a tie"
    #If the board is not full and there is no winner till now case
    
    else:
        return "This is an ongoing game"
```

```python
print(check_winner(board))
```
